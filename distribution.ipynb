```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(5, 3,figsize=(10, 10))
for idx, ax in enumerate(axes.flatten()):
    ax.set_title(f"sample size {np.power(2, idx)}")
    sns.histplot(np.random.chisquare(np.power(2, idx), 1000), ax=ax, kde=True)

fig.tight_layout()
plt.show()
```
