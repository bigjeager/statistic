```python
from matplotlib import pyplot
import numpy as np
import pandas as pd

pop = np.random.randint(1, 10000, 5000)

sample_size = []
sample_mean = []
for i in range(10, 10000, 10):
    sample_size.append(i)
    sample_mean.append(np.random.choice(pop, i, replace=True).mean())

pd.DataFrame({"sample_size": sample_size, "sample_mean": sample_mean}).plot(x='sample_size', y='sample_mean')
pyplot.axhline(pop.mean(), color='red')
```
